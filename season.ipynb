```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
movies = pd.read_csv('movies.csv') #open main df
```

```python
movies['release_month'].hist(figsize=(16,6), bins=12); #histogram by month of release
plt.title('Average Number of Movie Releases by Month', fontsize=20)
plt.ylabel('Average Number of Releases', fontsize=16)
plt.xlabel('Month', fontsize=16)
plt.show()
```

<b> The graph shows how many movies are released in those months, which is an indicator of competition on those months for ticket sales. </b>

```python
grouped = movies.groupby('release_month').sum() #bar graph of profitable factors by month
grouped.plot(kind='barh', figsize=(15,15))
plt.show();
```

```python
month_roi = pd.DataFrame() #create new df
```

```python
month_roi['ROI_avg'] = movies.groupby('release_month', sort=True)['ROI%'].mean() #avg ROI by month
```

```python
month_roi
```

<b> We now know that July and October offer the highest ROI on average, but higher competition based on the histogram</b>

```python
month_roi.plot(kind='bar', figsize=(10,5), color='bbbbbgbbbbgb') #bar graph of avg ROI by month
plt.title('Average ROI by Release Month', fontsize=25, fontweight='bold')
plt.xlabel('Release Month')
plt.ylabel('ROI %')
plt.xticks(rotation=45);
```
